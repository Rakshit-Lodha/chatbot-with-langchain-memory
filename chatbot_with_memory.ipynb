```python
from langchain.agents import create_agent
```

```python
from langchain_core.tools import tool
```

```python
from langgraph.checkpoint.memory import InMemorySaver
```

```python
checkpointer = InMemorySaver()
```

```python
from langchain.chat_models import init_chat_model
```

```python
def configuration_setup():

    model = init_chat_model(
        "gpt-3.5-turbo",
        temperature = 0.7
    )
    
    agent = create_agent(
        model = model,
        system_prompt = f""" Based on the user input, 
        you are the friend of the user, you're supposed to give empathetic replies to the user. """,
        checkpointer = checkpointer
    )
    
    config = {"configurable": {"thread_id": "user_1"}}

    return config
```

```python
def input_gather():
    configuration = configuration_setup()

    sample = input("Write something?")
    
    response = agent.invoke(
        {"messages": [{"role": "user", "content": sample}]},
        config = configuration
    )
    
    response = response['messages'][-1].content

    return response
```

```python
input_gather()
```

```python
def final_output():
    while True:
        response = input_gather()
        print(response)
        satisfied = input("Would you like to continue the conversation? (Yes - Y, No - N)")
        if satisfied == "Y":
            response
        elif satisfied == "N":
            print("Okay, bye for now!")
            break
        else:
            print("Input the right value, Yes - Y or No - N")
```

```python
final_output()
```
