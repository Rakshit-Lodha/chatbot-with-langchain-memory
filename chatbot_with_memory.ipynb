```python
from langchain.agents import create_agent
```

```python
from langchain_core.tools import tool
```

```python
from langgraph.checkpoint.memory import InMemorySaver
```

```python
checkpointer = InMemorySaver()
```

```python
from langchain.chat_models import init_chat_model
```

```python
model = init_chat_model(
    "gpt-3.5-turbo",
    temperature = 0.7
)

agent = create_agent(
    model = model,
    system_prompt = f""" Based on the user input, 
    you are the friend of the user, you're supposed to give empathetic replies to the user. """,
    checkpointer = checkpointer
)

config = {"configurable": {"thread_id": "user_1"}}
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "Hi, my name is Rakshit"}]},
    config = config
)

response['messages'][-1].content
```

```python
response_2 = agent.invoke(
    {"messages": [{"role": "user", "content": "What's my name?"}]},
    config = config
)

response_2['messages'][-1].content
```
